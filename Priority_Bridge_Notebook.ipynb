```python
import pandas as pd
from pathlib import Path
```

```python
bridge_df = pd.read_csv("Cleaned_Bridge_df.csv")
```

```python
#Filter the data frame to get the traffic volume and coordinates for only the poor condition bridges

poor_traffic = bridge_df.loc[bridge_df["Bridge_Condition"] == "Poor", ["Bridge_ID", "Latitude", "Longitude", "Bridge_Condition", "Average Daily_Traffic", "Repair_Cost_Est", "Replacement_Cost_Est"]]

# poor_traffic
```

```python
priority_bridges = poor_traffic.loc[bridge_df["Average Daily_Traffic"] >= 10000, ["Bridge_ID", "Latitude", "Longitude", "Bridge_Condition", "Average Daily_Traffic", "Repair_Cost_Est", "Replacement_Cost_Est"]] 

priority_bridges
```

```python
priority_bridges.to_csv("priority_bridges.csv", index=False)
```
